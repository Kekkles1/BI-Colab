# DATA WRANGLING

**Approach**
First we will drop unnnecessary columns, fill in the missing values into the important columns and then remove those rows where approximately 70% of the columns have missing data.

```python
#importing the basic libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import missingno as mano
%matplotlib inline
```

```python
#read the CSV file
df = pd.read_csv("hotel_bookings_dirty.csv")
```

```python
# remove arrival_date_year column and arrival_date_month  and 'arrival_date_weak_number and arrival_date_day_of_month and reservation_status and reservation_status_date column
df = df.drop(['arrival_date_year', 'arrival_date_month', 'arrival_date_week_number', 'arrival_date_day_of_month','reservation_status_date','deposit_type','previous_cancellations'], axis=1)
```

```python
df = df.drop(['babies', 'required_car_parking_spaces', 'total_of_special_requests','previous_bookings_not_canceled','booking_changes'], axis=1)
```

# CATEGORICAL (USED MODE)

```python
# Fill missing values in 'hotel' column with the mode
df['hotel'].fillna(df['hotel'].mode()[0], inplace=True)
```

```python
# Fill missing values in 'is_canceled' column using mode
df['is_canceled'].fillna(df['is_canceled'].mode()[0], inplace=True)
```

```python
# Fill missing values in 'is_repeated_guest' column using mode
df['is_repeated_guest'].fillna(df['is_repeated_guest'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

# NUMERICAL (MEAN/MEDIAN)

```python
from sklearn.impute import KNNImputer

# Create a KNN imputer with k=5
imputer = KNNImputer(n_neighbors=5)

# Select only the columns 'agent' and 'company' for imputation
columns_to_impute = ['agent', 'company']
df_to_impute = df[columns_to_impute]

# Perform KNN imputation
df_imputed = pd.DataFrame(imputer.fit_transform(df_to_impute), columns=columns_to_impute)

# Replace the original 'agent' and 'company' columns with the imputed values
df['agent'] = df_imputed['agent']
df['company'] = df_imputed['company']
```

```python
df.isnull().sum()
```

```python
# Select the columns for KNN imputation
columns_to_impute_knn = ['market_segment', 'distribution_channel', 'reservation_status']
df_to_impute_knn = df[columns_to_impute_knn].copy()  # Create a copy to avoid modifying the original DataFrame

# Convert categorical features to numerical using Label Encoding
from sklearn.preprocessing import LabelEncoder

for col in columns_to_impute_knn:
  le = LabelEncoder()
  df_to_impute_knn[col] = le.fit_transform(df_to_impute_knn[col].astype(str))

# Create a KNN imputer with k=5
imputer_knn = KNNImputer(n_neighbors=5)

# Perform KNN imputation
df_imputed_knn = pd.DataFrame(imputer_knn.fit_transform(df_to_impute_knn), columns=columns_to_impute_knn)

# Convert the imputed numerical values back to categorical using inverse transform
for col in columns_to_impute_knn:
  le = LabelEncoder()
  df_imputed_knn[col] = le.fit_transform(df_to_impute_knn[col].astype(str))
  df_imputed_knn[col] = le.inverse_transform(df_imputed_knn[col].astype(int))

# Replace the original columns with the imputed values
for col in columns_to_impute_knn:
  df[col] = df_imputed_knn[col]
```

```python
# Select the columns for KNN imputation including 'is_repeated_guest'
columns_to_impute_knn = ['market_segment', 'distribution_channel', 'reservation_status', 'is_repeated_guest']
df_to_impute_knn = df[columns_to_impute_knn].copy()  # Create a copy to avoid modifying the original DataFrame

# Convert categorical features to numerical using Label Encoding
for col in columns_to_impute_knn:
  if df[col].dtype == object:  # Check if the column is categorical
    le = LabelEncoder()
    df_to_impute_knn[col] = le.fit_transform(df_to_impute_knn[col].astype(str))

# Create a KNN imputer with k=5
imputer_knn = KNNImputer(n_neighbors=5)

# Perform KNN imputation
df_imputed_knn = pd.DataFrame(imputer_knn.fit_transform(df_to_impute_knn), columns=columns_to_impute_knn)

# Convert the imputed numerical values back to categorical using inverse transform
for col in columns_to_impute_knn:
  if df[col].dtype == object:  # Check if the column is categorical
    le = LabelEncoder()
    df_imputed_knn[col] = le.fit_transform(df_to_impute_knn[col].astype(str))
    df_imputed_knn[col] = le.inverse_transform(df_imputed_knn[col].astype(int))


# Replace the original columns with the imputed values
for col in columns_to_impute_knn:
  df[col] = df_imputed_knn[col]
```

```python
# Select the columns for KNN imputation including 'is_canceled' and 'reservation_status'
columns_to_impute_knn = ['is_canceled', 'reservation_status']
df_to_impute_knn = df[columns_to_impute_knn].copy()  # Create a copy to avoid modifying the original DataFrame

# Convert categorical features to numerical using Label Encoding
for col in columns_to_impute_knn:
  if df[col].dtype == object:  # Check if the column is categorical
    le = LabelEncoder()
    df_to_impute_knn[col] = le.fit_transform(df_to_impute_knn[col].astype(str))

# Create a KNN imputer with k=5
imputer_knn = KNNImputer(n_neighbors=5)

# Perform KNN imputation
df_imputed_knn = pd.DataFrame(imputer_knn.fit_transform(df_to_impute_knn), columns=columns_to_impute_knn)

# Convert the imputed numerical values back to categorical using inverse transform
for col in columns_to_impute_knn:
  if df[col].dtype == object:  # Check if the column is categorical
    le = LabelEncoder()
    df_imputed_knn[col] = le.fit_transform(df_to_impute_knn[col].astype(str))
    df_imputed_knn[col] = le.inverse_transform(df_imputed_knn[col].astype(int))


# Replace the original columns with the imputed values
for col in columns_to_impute_knn:
  df[col] = df_imputed_knn[col]
```

```python
# Fill missing values in 'adr' column using the median
df['adr'].fillna(df['adr'].median(), inplace=True)
```
